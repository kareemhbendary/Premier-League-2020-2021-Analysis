```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
#reading dataset
```

```python
epl_df=pd.read_csv("EPL_20_21.csv")
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
```

```python
epl_df.isna().sum()
```

```python
epl_df["MinsPerMatch"]= (epl_df["Mins"]/epl_df["Matches"]).astype(int)
epl_df["GoalsPerMatch"]= (epl_df["Goals"]/epl_df["Matches"]).astype(float)
epl_df.head()
```

```python
Total_Goals = epl_df["Goals"].sum()
Total_Goals
```

```python
Total_Penalty_Goals = epl_df["Penalty_Goals"].sum()
Total_Penalty_Goals
```

```python
Total_Penalty_Attempts = epl_df["Penalty_Attempted"].sum()
Total_Penalty_Attempts
```

```python
pl_not_scored = Total_Penalty_Attempts-Total_Penalty_Goals
data=[pl_not_scored,Total_Penalty_Goals]
labels=["Penalties Missed","Penalties Scored"]
color=sns.color_palette("Set2")
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.title("Total Penalties")
plt.show()
```

```python
clubs=epl_df.groupby("Club").size().sort_values(ascending=False)
clubs.head(7).plot(kind='bar',figsize=(10,6),color=sns.color_palette("viridis"))
plt.title("Number of players per club")
```

```python
epl_df.groupby("Age").size()
```

```python
under20= epl_df[epl_df['Age']<20]
age20_25= epl_df[(epl_df['Age']>=20) & (epl_df['Age']<26)]
age26_30= epl_df[(epl_df['Age']>=26) & (epl_df['Age']<30)]
above30= epl_df[epl_df['Age']>30]
```

```python
xunder20=under20["Name"].count()
xage20_25=age20_25["Name"].count()
xage26_30=age26_30["Name"].count()
xabove30=above30["Name"].count()
```

```python
data=[xunder20,xage20_25,xage26_30,xabove30]
labels=["Under 20", "20-25","26-30","30+"]
plt.pie(data,labels=labels,colors=sns.color_palette("Set2"),autopct='%.1f%%')
plt.title("Age Brackets")
plt.show()
```

```python
under20.groupby("Club").size().sort_values(ascending=0).plot(kind="bar",color=sns.color_palette("icefire"))
plt.title("Number of players under 20 per club")
```

```python
sns.boxplot(x="Club",y="Age",data=epl_df)
plt.title("Average age of players per club")
plt.xticks(rotation=90)
```

```python
teamsize=epl_df.groupby("Club").size()
data=(epl_df.groupby("Club")["Age"].sum())/teamsize
data.sort_values(ascending=0)
```

```python
assists_by_club= pd.DataFrame(epl_df.groupby('Club', as_index=0)["Assists"].sum())
sns.set_theme(style="whitegrid")
ax=sns.barplot(x="Club",y="Assists",data=assists_by_club.sort_values(by="Assists"),palette="rocket")
ax.set_xlabel("Club")
ax.set_ylabel("Number of assists")
plt.title("Number of assists per club")
plt.xticks(rotation=90)
```

```python
top_10_assisters=epl_df[["Name","Club","Assists"]].nlargest(n=10,columns="Assists")
top_10_assisters
```

```python
goals_by_club= pd.DataFrame(epl_df.groupby('Club', as_index=0)["Goals"].sum())
sns.set_theme(style="whitegrid")
ay=sns.barplot(x="Club",y="Goals",data=goals_by_club.sort_values(by="Goals"),palette="rocket")
ay.set_xlabel("Club")
ay.set_ylabel("Number of Goals")
plt.title("Number of goals per club")
plt.xticks(rotation=90)
```

```python
top_10_goalers=epl_df[["Name","Club","Goals"]].nlargest(n=10,columns="Goals")
top_10_goalers
```

```python
top_10_goals_per_match=epl_df[["Name","Club","Goals","Matches","GoalsPerMatch"]].nlargest(n=10,columns="GoalsPerMatch")
top_10_goals_per_match
```

```python
Total_assists=epl_df["Assists"].sum()
data=[Total_Goals-Total_assists,Total_assists]
labels=["Goals without assists","Goals with assists"]
plt.pie(data,labels=labels,colors=sns.color_palette("Set2"),autopct='%.0f%%')
plt.title("Goals with and without assists")
plt.show()
```

```python
yellowcards=epl_df[["Name","Club","Yellow_Cards"]].nlargest(n=10, columns="Yellow_Cards")
plt.title("Players with the most yellow cards")
c=sns.barplot(x=yellowcards["Name"],y=yellowcards["Yellow_Cards"],label='Players',color='yellow')
plt.ylabel("Number of Yellow cards")
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c
```

```python
redcards=epl_df[["Name","Club","Red_Cards"]].nlargest(n=10, columns="Red_Cards")
plt.title("Players with the most Red cards")
r=sns.barplot(x=redcards["Name"],y=redcards["Red_Cards"],label='Players',color='red')
plt.ylabel("Number of Red cards")
r.set_xticklabels(c.get_xticklabels(),rotation=45)
r
```
